```python
import numpy as np
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
from PIL import Image
import copy
# path of the image
path = "./images/chaplin.jpg"
# load the image
image = Image.open(path).convert('L')
# convert image to numpy array
mat_image = np.array(image)
m, n = mat_image.shape
print(mat_image.shape)
imshow(mat_image,  cmap='gray', aspect='auto')
im = Image.fromarray(np.uint8(mat_image))
im.save("chaplin.jpg") 
```

```python
c = 2
mat_aux = np.zeros((c*m, c*n))
mat_resized = np.zeros((c*m, c*n))
for i in range (0, m):
    for j in range(0,n):
        for k in range(c):
             mat_aux[i][j*c + k] = mat_image[i][j]

for i in range (0, m):
    for j in range(0, c*n):
        for k in range(c):
             mat_resized[i*c+k][j] = mat_aux[i][j]


imshow(mat_resized,  cmap='gray', aspect='auto')

im = Image.fromarray(np.uint8(mat_resized))
im.save("chaplin_resized.jpg") 
```

```python
mat_T = mat_image.T
imshow(mat_T,  cmap='gray')
```
